## Base de datos principal
1. Título de la base de datos.
  * Nombre: Donación de Órganos en México.
  * Url: https://www.kaggle.com/joshdoit/master
  
  
2. Descripción de los datos.
  * ¿Qué tipo de datos son?  
    Es una tabla que contiene registros sobre el proceso de la donación de órganos.
  * Descripción de las columnas
    * SEXO: Texto corto. Muestra el sexo del donante, el cual puede ser femenino o masculino.
    * CODIGO_SEXO: Valor numérico entero, 1 se refiere a masculino y 2 a femenino.
    * TIPO_DONANTE: Texto corto. Muestra si es un donante vivo o muerto.
    * MUERTE: Texto corto. Muestra la causa de muerte, si no aplica es porque el paciente es de tipo vivo.
    * ENTIDAD_FEDERATIVA: Texto corto. Muestra el nombre de la entidad federativa donde se realiza el procedimiento médico.
    * CODIGO_ENTIDAD_FEDERATIVA: Valor numérico entero. Aquí se asigna un número a cada entidad federativa.
    * ESTABLECIMIENTO: Texto largo. Muestra el nombre del establecimiento donde se realiza el procedimiento médico.
    * INSTITUCION: Texto corto. Nombre de las instituciones que realizan el procedimiento de extracción de órganos
    * EDAD_ANIOS: Valor numérico entero. Muestra la edad del donante.
    * FECHA_PROCURACION: Tiempo (DataTime). Muestra la fecha en la que se realiza el procedimiento médico.
    * RINON_IZQUIERDO: Valor numérico entero. Cantidad de riñones izquierdos donados.
    * RINON_DERECHO: Valor numérico entero. Cantidad de riñones derechos donados.
    * RINON_BLOCK: Valor numérico entero. Cantidad de bloques de riñones donados.
    * PULMON_IZQUIERDO: Valor numérico entero. Cantidad de pulmones izquierdos donados.
    * PULMON_DERECHO: Valor numérico entero. Cantidad de pulmones derechos donados.
    * CORAZON: Valor numérico entero. Cantidad de corazones donados.
    * HIGADO: Valor numérico entero. Cantidad de hígados donados.
    * PANCREAS: Valor numérico entero. Cantidad de páncreas donados.
    * INTESTINO: Valor numérico entero. Cantidad de intestinos donados.
    * CORNEA_IZQUIERDA: Valor numérico entero. Cantidad de corneas izquierdas donadas.
    * CORNEA_DERECHA: Valor numérico entero. Cantidad de corneas derechas donadas.
    * PIEL: Valor numérico entero. Cantidad de veces que ha donado piel.
    * HUESOS: Valor numérico entero. Cantidad de huesos donados.
    * CORAZON_TEJIDOS: Valor numérico entero. Cantidad de tejidos de corazón donados.
